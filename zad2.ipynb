Zadanie to polega na znalezieniu równania prostej opisującej dane ekperymentalne. Zostało wykonane z pomocą pakietu numpy w języku python

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

# kwintesencja metody najmnijeszych kwadratów
def leastSquares(x,y):
    a = np.vstack([x, np.ones(len(x))]).T
    return np.dot(np.linalg.inv(np.dot(a.T, a)), np.dot(a.T, y))

# dane wejściowe
X = [1,3,5] #[N]
Y= [0,0.4,1.15] #[mm]

linear_model = leastSquares(x=X,y=Y)

# wizualizacja
plt.scatter(X,Y,color="red",label="Zebrane punkty")
line=[linear_model[0]*i+linear_model[1] for i in X]
plt.plot(X,line,color="blue",label=f"aproksymacja liniowa y={linear_model[0]:.4f}x + {linear_model[1]:.4f}")
plt.legend()
# plt.rcParams['figure.figsize'] = [10, 5]
plt.xlabel("obciążenie [N]")
plt.ylabel("długość [mm]")
plt.show()
```
